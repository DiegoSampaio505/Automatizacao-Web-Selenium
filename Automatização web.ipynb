```python
#!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
navegador = webdriver.Chrome()#abrir navegador

navegador.get('https://www.google.com.br')#acessar o site

navegador.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("dolar", Keys.ENTER)#Find element foi usado para selecionar a barra de pesquisa do google, send_keys foi usado para escrever dolar e keys.enter foi usado para precionar o enter

dolar = navegador.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')#aqui criei a variavel para cotação do dolar, o processo foi o mesmo, o get_attribute foi usado para definir o que eu queria atribur a variavel dentro do XPATH
```

```python
navegador.get('https://www.google.com.br')#acessar o site

navegador.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("euro", Keys.ENTER)#Find element foi usado para selecionar a barra de pesquisa do google, send_keys foi usado para escrever dolar e keys.enter foi usado para precionar o enter

euro = navegador.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')#aqui criei a variavel para cotação do euro, o processo foi o mesmo, o get_attribute foi usado para definir o que eu queria atribur a variavel dentro do XPATH
```

```python
navegador.get('https://www.melhorcambio.com/ouro-hoje')

ouro = navegador.find_element(By.XPATH,'//*[@id="comercial"]').get_attribute('value')#mesmo procedimento

navegador.quit()

print(dolar, euro, ouro)
```

```python
ouro = ouro.replace(",", ".") # transformando a cotação do ouro em numero (numero em python é sempre separado por . e nao por ,)

dolar = dolar.replace(",", ".")

euro = euro.replace (",", ".")


print(ouro)
```

```python
tabela = pd.read_excel('Produtos.xlsx')
display(tabela)
```

```python
tabela.loc [tabela["Moeda"]=="Dólar","Cotação"] = float(dolar) #passei a cotação do dolar para as linhas da coluna moeda cujo nome eram dolar

tabela.loc [tabela["Moeda"]=="Euro","Cotação"] = float(euro)# passei a cotação do euro para as linhas da coluna moeda cujo nome eram euro

tabela.loc [tabela["Moeda"]=="Ouro","Cotação"] = float(ouro)#passei a cotação do ouro para a linha da coluna moeda cujo nome era ouro

display(tabela)
```

```python
tabela["Preço de Compra"] = tabela["Preço Original"] * tabela["Cotação"] #Fiz a multiplicação das tabela preço original e cotação para descobrir o preço atual de compra

tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem"] #Fiz a multiplicação das colunas preço de compra e margem para saber o preço de venda

tabela["Lucro"] = tabela["Preço de Venda"] - tabela["Preço de Compra"] #Criei a tabela de lucro com o valor da subtração entre o prçeo de venda e o preço de compra 
    
display(tabela)
```

```python
tabela.to_excel("Produtos.xlsx")#exportar tabela

tabela.to_excel("Produtos.xlsx", index=False) #exportar tabela sem index
```
